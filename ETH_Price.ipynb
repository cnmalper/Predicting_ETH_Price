```python
y = dataFrame["High"].values
x = dataFrame.drop("High",axis=1).values
# Convert to numpy series. 
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(4,activation="relu")) 
model.add(Dense(4,activation="relu")) 
model.add(Dense(4,activation="relu")) 
model.add(Dense(4,activation="relu")) 

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x = x_train, y = y_train, validation_data = (x_test, y_test), batch_size = 25, epochs = 300)
```

```python
lostData = pd.DataFrame(model.history.history)
```

```python
lostData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizi = model.predict(x_test)
```

```python
tahminDizi
```

```python
mean_absolute_error(y_test,tahminDizi) # hata farkÄ± (y_test ve tahminDizi arasÄ±nda)
```
